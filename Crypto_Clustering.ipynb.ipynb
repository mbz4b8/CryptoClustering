```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler to the data and transform the data
data_scaled = scaler.fit_transform(df_market_data)
data_scaled
```

```python
# Create a DataFrame with the scaled data
data_scaled_df = pd.DataFrame(data_scaled, columns=df_market_data.columns)

# Copy the crypto names from the original data
data_scaled_df['coin_id'] = df_market_data.index

# Set the coinid column as index
data_scaled_df.set_index('coin_id', inplace=True)

# Display sample data
data_scaled_df.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for i in k:
    # Create a KMeans model using the loop counter for the n_clusters
    k_model = KMeans(n_clusters=i, random_state=1, n_init=10)
    
    # Fit the model to the data using df_market_data_scaled
    k_model.fit(data_scaled_df)
    
    # Append the model.inertia_ to the inertia list
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {'k': k, 'inertia': inertia}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_df.hvplot.line(x="k", y="inertia", title="Elbow Curve")
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 4

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
k_model=KMeans(n_clusters=4, random_state=4)
```

```python
# Fit the K-Means model using the scaled data
k_model.fit(data_scaled_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
k_4=k_model.predict(data_scaled_df)

# Print the resulting array of cluster values.
k_4
```

```python
# Create a copy of the DataFrame
data_predictions_df = data_scaled_df.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
data_predictions_df['Prediction_Cluster'] = k_4

# Display sample data
data_predictions_df.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
data_predictions_df.hvplot.scatter(x="price_change_percentage_24h", y="price_change_percentage_7d",
                                                  by="Prediction_Cluster", hover_cols=["coin_id"])
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca_model = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
data_pca= pca_model.fit_transform(data_scaled_df)
# View the first five rows of the DataFrame. 
print(data_pca[:5])
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
explained_variance_ratio = pca_model.explained_variance_ratio_
print("Variance for each com")
# Calculate the total explained variance
total_explained_variance = sum(explained_variance_ratio)

print("Total Explained Variance of the three principal components:", total_explained_variance)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 89.5


```python
# Create a new DataFrame with the PCA data.


# Creating a DataFrame with the PCA data
pca_data_df = pd.DataFrame(data_pca, columns= ["PCA1", "PCA2", "PCA3"])

# Copy the crypto names from the original data
pca_data_df['coin_id'] = df_market_data.index

# Set the coinid column as index
pca_data_df.set_index('coin_id', inplace=True)

# Display sample data
pca_data_df.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
pca_k = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
pca_inertia = []

# Create a for loop to compute the inertia with each possible value of k
for i in pca_k:
    # Create a KMeans model using the loop counter for the n_clusters
    pca_k_model = KMeans(n_clusters=i, random_state=1, n_init=10)
    
    # Fit the model to the data using df_market_data_scaled
    pca_k_model.fit(pca_data_df)
    
    # Append the model.inertia_ to the inertia list
    pca_inertia.append(pca_k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
pca_elbow_data = {'pca_k' : k, 'pca_inertia': inertia
                 }
# Create a DataFrame with the data to plot the Elbow curve
pca_elbow_df = pd.DataFrame(pca_elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
pca_elbow_df.hvplot.line(x="pca_k", y="pca_inertia", title="PCA Elbow Curve")
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** 4


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No, best value for K is still 4

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
k_model=KMeans(n_clusters=4, random_state=4)
```

```python
# Fit the K-Means model using the PCA data
k_model.fit(pca_data_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
pca_k_4=k_model.predict(pca_data_df)

# Print the resulting array of cluster values.
pca_k_4
```

```python
# Create a copy of the DataFrame with the PCA data
data_predictpca_df = pca_data_df.copy()

# Add a new column to the DataFrame with the predicted clusters
data_predictpca_df['PCA_Cluster'] = pca_k_4

# Display sample data
data_predictpca_df.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
data_predictpca_df.hvplot.scatter(x="PCA1", y="PCA2",
                                                  by="PCA_Cluster", hover_cols=["coin_id"])
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
# Composite plot to contrast the Elbow curves
# Plotting the Elbow curves using hvPlot
first_curve = elbow_df.hvplot.line(x="k", y="inertia", label="Non PCA Transformed Data")
pca_curve = pca_elbow_df.hvplot.line(x="pca_k", y="pca_inertia", label="PCA Transformed Data")

# Combining both curves into a single composite plot
composite_plot = first_curve * pca_curve

# Adding labels and title
composite_plot = composite_plot.opts(xlabel='Number of Clusters (k)', ylabel='Inertia', title='Elbow Curve Comparison', 
                                     show_legend=True)

# Display the composite plot
composite_plot
```

```python
# Composite plot to contrast the clusters
# Plotting the clusters using hvPlot
first_clusters = data_predictions_df.hvplot.scatter(x="price_change_percentage_24h", y="price_change_percentage_7d", 
                                                    by="Prediction_Cluster", label="Non PCA Data", marker='o').opts(
                                                    xlabel='Price Change Percentage (24h)', ylabel='Price Change Percentage (7d)', show_legend=True)
pca_clusters = data_predictpca_df.hvplot.scatter(x="PCA1", y="PCA2", by="PCA_Cluster", label="PCA Data", marker='s').opts(
                                                xlabel='PCA1', ylabel='PCA2', show_legend=True)

# Combining both cluster plots into a single composite plot
composite_clusters = first_clusters + pca_clusters

# Adding title
composite_clusters = composite_clusters.opts(title='Cluster Comparison')

# Display the composite plot
composite_clusters
```
